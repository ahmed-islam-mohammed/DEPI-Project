```python
import pandas as pd
from sqlalchemy import create_engine

# اتصال بقاعدة البيانات
engine = create_engine('mssql+pyodbc://*$#*$#/bike_store?driver=SQL+Server+Native+Client+11.0')

# قراءة بيانات من الجدول التشغيلى
customers_df = pd.read_sql('SELECT * FROM vw_Customer', engine)

# إزالة القيم المكررة
customers_df.drop_duplicates(inplace=True)

# ملء القيم الفارغة (Null Values)
customers_df.fillna({'phone': 'Unknown', 'email': 'noemail@example.com'}, inplace=True)

# اتصال بقاعدة بيانات مستودع البيانات
dw_engine = create_engine('mssql+pyodbc://*$#*$#/bike_storeDW?driver=SQL+Server+Native+Client+11.0')

# تحميل البيانات إلى جدول الأبعاد في مستودع البيانات
customers_df.to_sql('DimCustomer', dw_engine, if_exists='append', index=False)
```

```python
import pandas as pd
from sqlalchemy import create_engine

# اتصال بقاعدة البيانات التشغيلية
engine = create_engine('mssql+pyodbc://*$#*$#/bike_store?driver=SQL+Server+Native+Client+11.0')

# اتصال بقاعدة بيانات مستودع البيانات
dw_engine = create_engine('mssql+pyodbc://*$#*$#/bike_storeDW?driver=SQL+Server+Native+Client+11.0')

# جلب آخر قيمة supplier_id في جدول DimSupplier
query_max_supplier_id = "SELECT ISNULL(MAX(supplier_id), 0) FROM DimSupplier"
max_supplier_id = pd.read_sql(query_max_supplier_id, dw_engine).iloc[0, 0]

# جلب السجلات الجديدة من الجدول التشغيلي
query_new_suppliers = f"SELECT * FROM vw_Supplier WHERE supplier_id > {max_supplier_id}"
new_suppliers_df = pd.read_sql(query_new_suppliers, engine)

# التحقق إذا كانت هناك بيانات جديدة
if not new_suppliers_df.empty:
    # إزالة القيم المكررة
    new_suppliers_df.drop_duplicates(inplace=True)

    # ملء القيم الفارغة
    new_suppliers_df.fillna({'contact_phone': 'Unknown', 'email': 'noemail@example.com'}, inplace=True)

    # تحميل البيانات الجديدة إلى جدول الأبعاد في مستودع البيانات
    new_suppliers_df.to_sql('DimSupplier', dw_engine, if_exists='append', index=False)
    print(f"{len(new_suppliers_df)} new record(s) inserted into DimSupplier.")
else:
    print("No new records to insert.")
```
